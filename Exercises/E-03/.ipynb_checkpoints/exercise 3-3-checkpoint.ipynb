**1. Write a function that draws a grid like the following:**

```
+ - - - - + - - - - +
|         |         |
|         |         |
|         |         |
|         |         |
+ - - - - + - - - - +
|         |         |
|         |         |
|         |         |
|         |         |
+ - - - - + - - - - +
```

*Hint:* to print more than one value on a line, you can print a comma-separated sequence of values:

```py
print('+', '-')
```

By default, print advances to the next line, but you can override that behavior and put a space at the end, like this: 

```py
print('+', end=' ')
print('-')
```

The output of these statements is `'+ -'`.
A print statement with no argument ends the current line and goes to the next line.

````python
def do_twice(fun):
    fun()
    fun()

def do_four(fun):
    do_twice(fun)
    do_twice(fun)

def beam():
    print(f"{' -'*4} +", end="")

def piller():
    print(f"{'  '*4} |", end="")

def beam_two():
    print('+', end="")
    do_twice(beam)
    print()

def piller_two():
    print('|', end="")
    do_twice(piller)
    print()

def pillers():
    do_four(piller_two)

def row():
    pillers()
    beam_two()

def grid():
    beam_two()
    do_twice(row)

grid()
````

**2. Write a function that draws a similar grid with four rows and four columns.**

Solution: http://thinkpython2.com/code/grid.py. 

Credit: This exercise is based on an exercise in Oualline, Practical C Programming, Third Edition, Oâ€™Reilly Media, 1997.

````python
def piller_four():
    print('|', end="")
    do_four(piller)
    print()

def beam_four():
    print('+', end="")
    do_four(beam)
    print()

def pillers():
    do_four(piller_four)

def row_four():
    pillers()
    beam_four()

def grid_4_4():
    beam_four()
    do_four(row_four)

grid_4_4()
````

## Refined Approach

````python
def grid_automated(width = 4, height = 4, rows = 2, columns = 2):
    """Created a customized grid of any size
    
    width: width of a cell
    height: height of a cell
    rows: number of rows
    columns: number of columns
    
    eg.: grid_automated(2, 2, 4, 4)
    """
    n_line = '\n'

    beam = f"+{(' -' * width + ' +') * columns}\n"
    piller = f"|{('  ' * width + ' |') * columns}\n"
    pillers = f"{piller * height}"
    
    line = pillers + beam
    grid = beam + line * rows
    
    print(grid)
    
grid_automated(2, 2, 4, 4)
grid_automated()

grid_automated(1, 1, 8, 8)
grid_automated(4, 2, 4, 3)
````
